```python
## sort the list without functions[13,456,76,888,9877,12,344,53,2], 

##define function
def sort_list(InputList):
    ##create empty list
    sorted_list = []
    listlength = len(InputList)
    ##begin while loop, run for total number of items in list
    for i in range(listlength):
        ## set lowest value to first item in list, 
        lowestvalue = InputList[0]
        ## loop through list, if number is lower than lowest value, set lowest value to that number
        for number in InputList:
            if number < lowestvalue:
                lowestvalue = number
                ##append lowest value to sorted list, remove lowest value from input list
        sorted_list.append(lowestvalue)
        InputList.remove(lowestvalue)
    return sorted_list

numbers = [13,456,76,888,9877,12,344,53,2]
sorted_numbers = sort_list(numbers)
print(sorted_numbers)
```
